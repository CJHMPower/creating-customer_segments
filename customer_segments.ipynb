### Practice: Outlier Detection
For any analysis, it is a very important step to detect the outliers in the data preprocessing process. The appearance of the outliers causes the values ​​to be tilted when these values ​​are taken into account. There is a lot of rule of thumb about how to define what is the outliers in a dataset. Here we will use [Tukey's method of defining outliers] (http://datapigtechnologies.com/blog/index.php/highlighting-outliers-in-your-data-with-the-tukey-method/): a * The outlier step * is defined as an interquartile range (IQR) of 1.5 times. A data point If a feature is included in a feature other than the IQR of the feature, the data point is identified as an exception point.

In the following code unit, you need to do the following:
 - Assign the value of the 25th quantile of the specified feature to `Q1`. Use `np.percentile` to do this.
 - Assign the value of the 75th locale of the specified feature to `Q3`. Similarly, use `np.percentile` to do this.
 - assigns the result of the exception order of the specified feature to `step`.
 - Optionally by removing the exception value by adding the index to the `outliers` list.


```python
# Find the extreme high or low point for every feature

for feature in log_data.keys():
    
    
    Q1 = np.percentile(log_data[feature],25)
    
    Q3 = np.percentile(log_data[feature],75)
    
    step=1.5(Q3-Q1)
    print (Q1-step,Q3+step)
    
    # display outlier
    print "Data points considered outliers for the feature '{}':".format(feature)
    display(log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))])
    
# the outlier indics to be removed
outliers  = [65,66,75,128,154]

good_data = log_data.drop(log_data.index[outliers]).reset_index(drop = True)
```

### Practice: Creating clusters
For different situations, some of the problems you need the number of clusters may be known. But in the case where the number of clusters is not known as a priori **, we can not guarantee that the number of clusters is optimal for this data because we have a structure (if any) for the data unclear. However, we can measure the quality of the cluster by calculating the ** contour factor ** at the midpoint of each cluster. The [contour coefficient] of the data point (http://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html) measures the similarity of it to the cluster assigned to it, 1 (not similar) to 1 (similar). ** The average ** contour factor provides us with a simple way to measure the quality of clustering.

In the next code unit, you will achieve the following functions:
 - use a clustering algorithm on `reduced_data` and assign the result to` clusterer`, and set `random_state` so that the result can be reproduced.
 - Use `clusterer.predict` to predict clusters for each point in` reduced_data` and assign the result to `preds` '.
 - Use the attribute value of the algorithm to find the cluster center and assign them to `values`.
 - Predict the category of each sample point in `pca_samples` and assign the result to` sample_preds` '.
 - Import the sklearn.metrics.silhouette_score package and calculate the contour coefficients of `reduced_data` relative to` preds`.
   - Assign the contour coefficients to `score` and output the results.

```python
# TODO：Implement clustering algorithm on the reduced data after dimension reduction
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
clusterer = KMeans(n_clusters=2,random_state=1).fit(reduced_data)


preds = clusterer.predict(reduced_data)
centers = clusterer.cluster_centers_
sample_preds =clusterer.predict(pca_samples)

# TODO：compute the mean silhouette coefficient
score = silhouette_score(reduced_data,preds)
print score
```

### 练习: Data reconstruction
Each of the clusters provided in the visualization image above has a center point. These centers (or average points) are not the ones that actually exist in the data, but are all * averages that predict the data points in the cluster. For creating a customer classification problem, the center of a cluster corresponds to the average user * of that category. Since this data is now downgraded and scaled to a certain extent, we can restore the cost of the user represented by applying a reverse conversion.

In the following code unit, you will achieve the following functions:
  - Use `pca.inverse_transform` to inverse center and store the result in 'log_centers'
  - Use `np.exp` to reverse` log_centers` and store the result in `true_centers` '.


```python
# TODO：inverse centers
log_centers =pca.inverse_transform(centers)

# TODO：perform log-transformation to centers
true_centers = np.exp(log_centers)

# display the real center
segments = ['Segment {}'.format(i) for i in range(0,len(centers))]
true_centers = pd.DataFrame(np.round(true_centers), columns = data.keys())
true_centers.index = segments
display(true_centers)
```
